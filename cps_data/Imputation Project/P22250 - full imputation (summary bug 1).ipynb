## Imports and versions

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
sm.show_versions()
```

## Data

```python
puf = pd.read_csv('puf2011.csv')

# Create categorical sign column

puf['sign'] = np.where(puf['P22250'] == 0, 1, np.where(puf['P22250'] > 0, 2, 0))

predictors =  ['DSI', 'EIC', 'MARS', 'XTOT', 'E00200', 'E00300', 'E00400','E00600', 
               'E00650', 'E00800', 'E00900', 'E01100', 'E01400', 'E01500', 'E01700',
               'E02100', 'E02300', 'E02400', 'E03150', 'E03210', 'E03240', 'E03270',
               'E03300', 'E17500', 'E18400', 'E18500', 'E19200', 'E19800', 'E20100',
               'E20400', 'E32800', 'F2441', 'N24']

# Prune puf to include only: RECID, P22250, categorical and predictor columns. Fill NaN's with 0.

keep = ['RECID', 'P22250', 'sign'] + predictors

puf = puf[keep].fillna(0)
```

## Model

```python
# Formula = sign ~ DSI + EIC + ... + F2441 + N24

formula = 'sign ~ ' + ' + '.join(predictors)
model = smf.mnlogit(formula = formula, data= puf).fit()
```

```python
model.summary()
```

```python
# Producing a similar problem using iris dataset, MNLogit.
# Source = https://stackoverflow.com/questions/31507396/mnlogit-in-statsmodel-returning-nan

iris = sm.datasets.get_rdataset('iris','datasets')
y = iris.data.Species
x = iris.data.ix[:, 0:4]
x = sm.add_constant(x, prepend = False)
mdl = sm.MNLogit(y, x)
mdl_fit = mdl.fit()
print (mdl_fit.summary())
```
