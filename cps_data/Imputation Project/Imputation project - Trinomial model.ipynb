```r
library('nnet')
```

```r
puf2011 <- read.csv('puf2011clean.csv')

#Change dummy variables to factors (categorical)
puf2011$E02000_dum <- as.factor(puf2011$E02000_dum)
puf2011$E26270_dum <- as.factor(puf2011$E26270_dum)
puf2011$P23250_dum <- as.factor(puf2011$P23250_dum)
puf2011$P22250_dum <- as.factor(puf2011$P22250_dum)
```

```r
#80% training data, 20% validation
sample <- sample(2, nrow(puf2011), replace = TRUE, prob = c(0.8,0.2))
training <- puf2011[sample == 1,]
validation <- puf2011[sample == 2,]

 
```

```r
#Create models
E02000_model <- multinom(E02000_dum ~ DSI + EIC + MARS1 + MARS2 + MARS3 + MARS4 +
                                      E00200 + E00300 + E00600 + E00800 + E00900 +
                                      E01100 + E01400 + E01500 + E01700 + E02100 + 
                                      E02300 + E02400 + E03150 + E03210 + E03240 +
                                      E03270 + E03300 + E17500 + E18400 + E18500 + 
                                      E19200 + E19800 + E20100 + E20400 + E32800, data = training, maxit = 500)

E26270_model <- multinom(E26270_dum ~ DSI + EIC + MARS1 + MARS2 + MARS3 + MARS4 +
                                      E00200 + E00300 + E00600 + E00800 + E00900 +
                                      E01100 + E01400 + E01500 + E01700 + E02100 + 
                                      E02300 + E02400 + E03150 + E03210 + E03240 +
                                      E03270 + E03300 + E17500 + E18400 + E18500 + 
                                      E19200 + E19800 + E20100 + E20400 + E32800, data = training, maxit = 500)

P23250_model <- multinom(P23250_dum ~ DSI + EIC + MARS1 + MARS2 + MARS3 + MARS4 +
                                      E00200 + E00300 + E00600 + E00800 + E00900 +
                                      E01100 + E01400 + E01500 + E01700 + E02100 + 
                                      E02300 + E02400 + E03150 + E03210 + E03240 +
                                      E03270 + E03300 + E17500 + E18400 + E18500 + 
                                      E19200 + E19800 + E20100 + E20400 + E32800, data = training, maxit = 500)

P22250_model <- multinom(P22250_dum ~ DSI + EIC + MARS1 + MARS2 + MARS3 + MARS4 +
                                      E00200 + E00300 + E00600 + E00800 + E00900 +
                                      E01100 + E01400 + E01500 + E01700 + E02100 + 
                                      E02300 + E02400 + E03150 + E03210 + E03240 +
                                      E03270 + E03300 + E17500 + E18400 + E18500 + 
                                      E19200 + E19800 + E20100 + E20400 + E32800, data = training, maxit = 500)
```

```r
#Add prediction columns to validation dataframe
validation$E02000p_dum <- predict(E02000_model, validation)
validation$E26270p_dum <- predict(E26270_model, validation)
validation$P23250p_dum <- predict(P23250_model, validation)
validation$P22250p_dum <- predict(P22250_model, validation)

write.csv(validation, 'valid.csv')
```
